```python
# Re-import necessary libraries after kernel reset
import matplotlib.pyplot as plt

# Frameworks
frameworks = ["AutoGen", "AutoGPT", "Dify", "Semantic Kernel"]

# Scoring dimensions (05 scale)
# Model Integration: (flexibility + switching + heterogeneous)
# Tools Integration: (plugin support + custom tool support)

model_integration_scores = {
    "AutoGen": 3.5, 
    "AutoGPT": 2,
    "Dify": 4,
    "Semantic Kernel": 4
}

tools_integration_scores = {
    "AutoGen": 3,
    "AutoGPT": 1.8,
    "Dify": 3.3,
    "Semantic Kernel": 3.5
}

# Plot settings
fig, ax = plt.subplots(figsize=(7, 6))
for fw in frameworks:
    x = model_integration_scores[fw]
    y = tools_integration_scores[fw]
    ax.scatter(x, y, label=fw, s=270)
    ax.text(x + 0.05, y, fw, fontsize=12, ha='right', va='center')

# Axes and grid
ax.set_xlim(0, 5)
ax.set_ylim(0, 5)
ax.axhline(2.5, color='grey', linestyle='--', linewidth=1)
ax.axvline(2.5, color='grey', linestyle='--', linewidth=1)
ax.set_xlabel("Model Integration (05)", fontsize=16)
ax.set_ylabel("Tools Integration (05)", fontsize=16)
ax.set_title("Model vs. Tools Integration by Framework", fontsize=14)
# ax.grid(True, linestyle='--', alpha=0.6)

# Quadrant labels
label_color = 'grey'
ax.text(4.8, 4.9, "High Model & Tools", fontsize=12, ha='right', va='top', color=label_color)
ax.text(0.5, 4.9, "Low Model, High Tools", fontsize=12, ha='left', va='top', color=label_color)
ax.text(0.5, 2, "Low Model & Tools", fontsize=12, ha='left', va='bottom', color=label_color)
ax.text(4.8, 2, "High Model, Low Tools", fontsize=12, ha='right', va='bottom', color=label_color)

plt.tight_layout()
plt.show()
```
